# 07 — Generate Blog Figures

Load experiment CSVs from notebooks 02-06 and generate the final publication-ready dark-theme figures for the blog.

**No training.** Just loading CSVs and producing PNGs.

**Expected runtime:** ~2 minutes

```python
from pathlib import Path
import shutil

ROOT = Path("..").resolve()
FIGURES = ROOT / "figures"
BLOG_IMAGES = ROOT / "blog" / "images"
BLOG_IMAGES.mkdir(exist_ok=True)

# List all notebook-generated figures
nb_figures = sorted(FIGURES.glob("nb*_*.png"))
print(f"Found {len(nb_figures)} notebook figures:\n")
for f in nb_figures:
    size_kb = f.stat().st_size / 1024
    print(f"  {f.name:45s}  {size_kb:>7.0f} KB")

# Copy to blog/images/ for embedding in the HTML
print(f"\nCopying to {BLOG_IMAGES}/...")
for f in nb_figures:
    shutil.copy2(f, BLOG_IMAGES / f.name)
    
print("Done. Blog figures ready.")
```
